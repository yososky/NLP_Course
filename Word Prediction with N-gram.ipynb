```python
import bs4 as bs  # BeautifulSoup
import pandas as pd
import urllib.request
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.util import ngrams
from random import seed
from random import random
```

```python
#get text
def _scrape_webpage_fixed(fixed, url):
    """
    Use BeautifulSoup to scrape the webpage text contents.
    """    
    scraped_textdata = urllib.request.urlopen(url)
    textdata = scraped_textdata.read()
    parsed_textdata = bs.BeautifulSoup(textdata,'lxml')
    #paragraphs = parsed_textdata.find_all(['p', 'dd','li'])  #dd and li includes some informations, but some unrealted
    paragraphs = parsed_textdata.find_all(['p'])
    formated_text = ""
    
    if fixed == 0:
        for para in paragraphs:
            formated_text += para.text
    else:
        for para in paragraphs:
            line = para.text.lower() # change to lower case
            line = re.sub(r"[^a-z\s]", "", line.strip())  # only contains english words
            formated_text += line
    return formated_text
raw_text = _scrape_webpage_fixed(0 , 'https://en.wikipedia.org/wiki/Natural_language_processing')

#text
raw_text[0:300]
```

```python
stopword = stopwords.words('english')
from string import punctuation
def cleanUpSentense(token_sent):
    senttoken = nltk.word_tokenize(token_sent)
    #leSent = lemmatizeWords(senttoken)  # lemmatize senstnce
    noStopwords = [word for word in senttoken if word not in stopword]  # remove stop words
    formated_text = ""
    
    for para in noStopwords:
        line = para.lower() # change to lower case
        line = line.replace('\n','')
        line = line.replace('”','')
        line = line.replace('“','')
        outNumline = ''.join(c for c in line if not c.isdigit())  # take off number
        outPunc = ''.join(c for c in outNumline if c not in punctuation) # take off puncctutation
        formated_text += outPunc + " "
    return formated_text

fixed_text = cleanUpSentense(raw_text)
fixed_text[0:300]
```

## 2.1 (0.6 point) Follow code examples to compute Unigram & Bigram tables using FreqDist

```python
fixed_tokenized_text = nltk.word_tokenize(fixed_text)
unigram = nltk.FreqDist(fixed_tokenized_text)
unigram
```

```python
df = pd.DataFrame(unigram.most_common(),columns = ['Word','frequency'])
sorted_unigram_df = df.transpose() 
sorted_unigram_df
```

```python
def get_ngrams(text, N):
    text_token = nltk.word_tokenize(text.lower())
    n_grams = ngrams(text_token, N)
    return [" ".join(grams) for grams in n_grams]
```

```python
bigram_token = get_ngrams(fixed_text, 2)
bigram = nltk.FreqDist(bigram_token)
bigram
```

```python
df = pd.DataFrame(bigram.most_common(),columns = ['Word','frequency'])
sorted_bigram_df = df.transpose() 
sorted_bigram_df
```

## 2.2 (2 points) Write a function to calculate the relative frequency table

```python
bigramlist = []
for word in bigram:
    splits = word.split(' ')
    bigramlist.append(splits)
bigramlist[0:5]
```

```python
allwords = sorted_unigram_df.iloc[0].tolist()
allwords[0:5]
```

```python
for i in range(0, 437):
    divide = unigram[allwords[i]]
    for j in range(1, 438):            
        if dftest.iloc[i,j] == 0:
                continue;
        value = dftest.iloc[i][j] / divide
        dftest.iloc[i, j] = value
```

## 2.3 (1 point) Normalize each row in the relative frequency table

```python
normal_bigramtable = df.copy()
```

```python
v = len(allwords)

for i in range(0, 437):
    zeroevalue = 1.0 / (unigram[allwords[i]] + v)
    for j in range(1, 438):
        divide = unigram[allwords[j-1]]
                
        if normal_bigramtable.iloc[i,j] == 0:
            normal_bigramtable.iloc[i, j] = zeroevalue;
        else:
            value = (normal_bigramtable.iloc[i][j] + 1) / (divide + v)
            normal_bigramtable.iloc[i, j] = value
```

## 2.4 (3 points) Write a function for Word Prediction

2.4.1 Input a word

2.4.2 Choose a random number r (0<r<1) to see if it falls on the distribution

2.4.3 If yes, output the predicted word; else output “NULL”.

```python
def predict_next_word(firstword, rand):
    row = bigramtable.loc[bigramtable['first/second'] == firstword]
    sumvalue = 0
    value_dict = {}
    for (columnName, columnData) in row.items():
        if columnName == 'first/second':
            continue
        else:
            value_dict[columnName] = columnData.values[0]
            sumvalue += columnData.values[0]
    pairs = sorted(value_dict.items(), key=lambda item: item[1], reverse= True)
    res = list()
    for pair in pairs:
        if pair[1] <= 0:
            continue
        res.append(pair)
    for i, pair in enumerate(res):
        name = res[i][0]
        value = res[i][1]/sumvalue
        tmp = [name, value]
        res[i] = tmp

    for value in res:
        rand -= value[1]
        if rand <= 0:
            return value[0]
            break
```

```python
def predict_word(firstword, rand):
    if firstword in allwords:
        return predict_next_word(firstword, rand)
    else:
        return 'null'
```

```python
def main():
    word = input('Type a word: ')  #input a word
    ran_val = random()  # random a value 0<r<1
    print('The next word is: ' + str(predict_word(word, ran_val)))
if __name__ == "__main__":
    main()
```
